```python
#Python
#Numpy

#Numpy is a library for scientific computing
#In this notebook, we will be covering numpy in 1D, in particular "n d arrays"
# we will be covering:
# 1) The Basics and Array Creation 
#2) Indexing and Slicing 
#3) Basic Operations 
#4) Universal Functions

#List
#Let's go over how to create a numpy array
#A Python list is a container that allows you to store and access data
print('\n A usual list: ')
PythonList1=['0', 1, 'two', '3', 4]
# Each element is associated with an index
# We can access each element using a square bracket, as follows:
print(PythonList1[0])
print(PythonList1[1])
print(PythonList1[2])
print(PythonList1[3])


#"numpy" array or "ndarray"
#A "numpy" array or "ndarray" is similar to a list
#It's usually fixed in size and each element is of the same type, in this case, integers
[1,2,3,4]

#CAST A LIST TO A NUMPY ARRAY
#We can cast a list to a numpy array by first importing numpy
print('\n Casting a list to a numpy array:')

import numpy as np
NUMPY_ARRAY=np.array([1,2,3,4])
print(NUMPY_ARRAY)
# We then cast the list as follows
# We can access the data via an index
# As with a list, we can access each element with an integer and a square bracket
print('\n access each element in NumPy:')
print(NUMPY_ARRAY[0])
print(NUMPY_ARRAY[1])
print(NUMPY_ARRAY[2])
print(NUMPY_ARRAY[3])


# If we check the type of the array we get "numpy.ndarray"
print('\n Checking the type of the array we get "numpy.ndarray"')
print(type(NUMPY_ARRAY))

# As numpy arrays contain data of the same type
# We can use the attribute "dtype" to obtain
print('\n Checking the type of numpy arrays contain data of the same type')
print( NUMPY_ARRAY.dtype)

#The attribute size is the number
NUMPY_ARRAY2=np.array([1,2,3,4, 4, 6, 8 , 1])
print('\n size Ex1:')
print(NUMPY_ARRAY.size)
print('\n size Ex2:')
print(NUMPY_ARRAY2.size)

#The attribute "ndim” represents the number of array dimensions or the rank of the array in this case one.
print('\n The number of array dimensions Ex2:')
print(NUMPY_ARRAY2.ndim)
NUMPY_ARRAY3=np.array([2,3 ,4, 5, 67,3, 2 , 5])
print('\n The number of array dimensions Ex3:')
print(NUMPY_ARRAY3.ndim)

#The attribute "shape” is a tuple of integers indicating the size of the array in each dimension
print('\n The attribute "shape” is a SIZE tuple of integers Ex3:')
print(NUMPY_ARRAY3.shape)

#The attribute "shape” is a tuple of integers indicating the size of the array in each dimension
print('\n The attribute "shape” is a SIZE tuple of integers Ex3:')
print(NUMPY_ARRAY3.shape)
```

```python
#Numpy with real number
#We can create a numpy array with real numbers
#When we check the type of the array, we get numpy.ndarray.

import numpy as Behdadnm
NUMPY_ARRAY4=Behdadnm.array([23.7, 298.67, 45, -0.004, 12, 15, 19.003, 231])
print(NUMPY_ARRAY4)

#type
print('\n Type of numpy array:')
print(type(NUMPY_ARRAY4))
#If we examine the attribute "dtype," we see float 64 as the elements are not integers
print('\n the attribute dtype to eextract the type of numpy arry:')
print(NUMPY_ARRAY4.dtype)

#Let's review some Indexing and Slicing methods
#We can change the first element of the array to 400, as follows:
print("\n NUMPY_ARRAY4 before replacing the first element:", NUMPY_ARRAY4)
NUMPY_ARRAY4[0]=400
print("\n NUMPY_ARRAY4 after replacing the first element:", NUMPY_ARRAY4)

#We can change the 6th element of the array as follows
#The sixth element is now 300
NUMPY_ARRAY4[5]=300
print("\n NUMPY_ARRAY4 after replacing the 6th element:", NUMPY_ARRAY4)
```

```python
#Slice
#Like lists and tuples, we can slice a numpy array 
import numpy as tigo
NUMPY_ARRAY5=tigo.array([23.7, 2.7, 45, -0.4, 12, 15, 19.3, 23, 65, 2, 3])
print(NUMPY_ARRAY5)

nMp1=NUMPY_ARRAY5[0:3]
print(nMp1)

nMp2=NUMPY_ARRAY5[1:5]
print(nMp2)

nMp3=NUMPY_ARRAY5[4:8]
print(nMp3)


#Like lists and tuples, we can slice a numpy array
#The elements of the array correspond to the following index: 0, 1, 2, ..., n-1, n is the number of elements

#We can assign the corresponding indexes to new values as follows:
NUMPY_ARRAY5[4:7]=201,301,401
print('\nNew NUMPY_ARRAY5: \n', NUMPY_ARRAY5)

NUMPY_ARRAY5[4:7]=201, 1111,401
print('\nNew NUMPY_ARRAY5: \n', NUMPY_ARRAY5)

NUMPY_ARRAY5[4:7]=201,301,99999
print('\nNew NUMPY_ARRAY5: \n', NUMPY_ARRAY5)
```

```python
#Numpy
#Basic Operations

#Numpy makes it easier to do many operations that are commonly performed in data science
#These same operations are usually computationally faster and require less memory in numpy compared


#Vector addition
#Vector addition is a widely used operation in data science

#Consider the vector 'u' with two elements;
#the elements are distinguished by the different colors
#Similarly, consider the vector 'v' with two components
#In vector addition, we create a new vector in this case 'z'
#The first component of 'z' is the addition of the first component of vectors 'u' and'v'
#Similarly, the second component is the sum of the second components of 'u' and 'v'
vector1=[1, 2]
vector2=[6, 2]
print('Vector 1:\n', vector1)
print('\nVector 2:\n', vector2)
#The following 3 lines of code will add the two lists and place the result in the list 'z'
z=[]
for n,m in zip(vector1, vector2):
    z.append(n+m)
    
print('\n Result vector 1 + vector 2= Vector:')
print(z)

vector1=[1, 2, 20, 71, 0, -7]
vector2=[4, 5, 25, 51, 5, 33]
print('Vector 1:\n', vector1)
print('\nVector 2:\n', vector2)
#The following 3 lines of code will add the two lists and place the result in the list 'z'
z=[]
for n,m in zip(vector1, vector2):
    z.append(n+m)
    
print('\n Result vector 1 + vector 2= Vector:')
print(z)


#We can also perform vector addition with one line of numpy code
import numpy as vect
vector1=vect.array([1, 2, 20, 71, 0, -7])
vector2=vect.array([4, 5, 25, 51, 5, 33])
print('\n Result vector 1 + vector 2= Vector:')
z=vector1+vector2
print(z)


#Subtraction 
#We can also perform vector subtraction by changing the addition sign to a subtraction sign
print('Subtraction')
vector1=[1, 2, 20, 71, 0, -7]
vector2=[4, 5, 25, 51, 5, 33]
print('Vector 1:\n', vector1)
print('\nVector 2:\n', vector2)
#The following 3 lines of code will add the two lists and place the result in the list 'z'
z=[]
for n,m in zip(vector1, vector2):
    z.append(n-m)
print('\n Result vector 1 - vector 2= Vector:')
print(z)

vector1=vect.array([1, 2, 20, 71, 0, -7])
vector2=vect.array([4, 5, 25, 51, 5, 33])
print('\n Result vector 1 - vector 2= Vector:')
z=vector1-vector2
print(z)
```

```python
#Numpy
#Vector multiplication
#Vector multiplication with a Scalar is another commonly performed operation
import numpy as mult

vector1=mult.array([134, 234, 203, 571, 40, -37])
vector2=mult.array([4, 5, 25, 51, 5, 33])

print('\nVector multiplication Ex1:')
z=2*vector1
print(z)

print('\nVector multiplication Ex2:')
z=3*vector2
print(z)

#Another Method for multiplication
z=[]
print('\nAnother Method for multiplication:')
for n in vector1:
    z.append(2*n)
print(z)


#Hadamard product
#Hadamard product is another widely used operation in data science
print('\nHadamard product:','\n Z= Vector1*vector2:')
z=vector1*vector2
print(z)



#Another way for Hadamard product
vector1=mult.array([134, 234, 203, 571, 40, -37])
vector2=mult.array([4, 5, 25, 51, 5, 33])
z=[]
print('\nAnother Method for multiplication:')
for n,m in zip(vector1, vector2):
    z.append(n*m)
print(z)


#Dot Product
print('Dot Product')
vector1=mult.array([134, 234, 203, 571, 40, -37])
vector2=mult.array([4, 5, 25, 51, 5, 33])
print('\nDot Product Result:')
z=mult.dot(vector1,vector2)
print(z)
```

```python
#numpy
#universal functions
#We can apply a universal function to a numpy array
#We can calculate the mean or average value of all the elements in 'a' using the method "mean”
import numpy as np
Numarray1=np.array([2,3,-2,1])

Mean_Numarray1=Numarray1.mean()
print(Numarray1)
print('Mean:', Mean_Numarray1)
```
